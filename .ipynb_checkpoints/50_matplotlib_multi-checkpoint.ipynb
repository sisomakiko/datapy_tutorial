# 複数のグラフ（図）を描画する

1. plt.subplot()を使う

2-1. plt.figure()を使う

2-2. plt.subplots()を使う

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

## plt.subplot()

```python
x = np.linspace(-10, 10, 100)
plt.subplot(1, 2, 1)
plt.plot(x, x**2)

plt.subplot(1, 2, 2)
plt.plot(x, x**3)
```
