```python
# カラムが数字でindexが文字列の場合
ndarray = np.arange(25).reshape(5, 5)
index = ['a', 'b', 'c', 'd', 'e']
columns = [0, 1, 2, 3, 4]
df = pd.DataFrame(ndarray, index=index, columns=columns)
df
```

```python
df[0]
```

```python
# 行のindexが文字列の場合
df.loc['a']
```

```python
s = pd.Series([100, 100, 100, 100, 100], index=['a', 'b', 'c', 'd', 'e'])
df[0] = s
df
```

```python
df[5] = s
df
```
