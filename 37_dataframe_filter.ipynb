# Pandas

```python
import pandas as pd
import numpy as np
```

## Filter

```python
data = {
    'name':['John', 'Zack', 'Emily'],
    'sex':['male', 'male', 'female'],
    'age':[22, 30, 32]
}
df = pd.DataFrame(data)
df
```

```python
# Trueの行だけ抽出される
df[[True, False, True]]
```

```python
filter_series = pd.Series([True, False, True])
filter_series
```

```python
df[filter_series]
```

```python
filter_s = df['age'] >= 30
df[filter_s]
```

```python
df = pd.read_csv('archive/tmdb_5000_movies.csv')
df[df['original_language'] == 'ja']
df[df['original_language'] == 'ja'].describe()
```

```python
df.describe()
```

```python
df[(df['original_language'] == 'ja') & (df['vote_average'] > 8)]
```

```python
# 予算が0　もしくは　売上が0
df[(df['budget'] == 0) | (df['revenue'] == 0)]
```

## ~でNOT演算

```python
# 予算も売上も0ではない
df[~((df['budget'] == 0) | (df['revenue'] == 0))]
```
