## .dropna() 

```python
# NaNを含むレコードを排除する
df.dropna().head()
```

```python
df[np.isnan(df['runtime'])]
```

```python
# あるカラムを指定してnanのレコードを排除する
# subset=[]で指定
df.dropna(subset=['runtime']).head()
```

# .fillna()

```python
df = pd.read_csv('archive/tmdb_5000_movies.csv')
df.head()
```

```python
df['runtime'].mean()
```

```python
# nanのところに平均値を入れる
df['runtime'] = df['runtime'].fillna(df['runtime'].mean())
```

```python
# nanがないか確認
df[np.isnan(df['runtime'])]
```
