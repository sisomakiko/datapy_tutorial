# groupby

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('archive/tmdb_5000_movies.csv')
df.head()
```

```python
df.groupby('original_language').mean()
```

```python
df.groupby('original_language').mean().loc['ja']
```

```python
df[df['original_language'] == 'ja']['budget']
```

```python
df[df['budget'] != 0].groupby('original_language').mean()
```

```python
df[df['budget']==0].groupby('original_language').count()
```

```python
df.groupby('original_language').describe()
```

```python
df.groupby('original_language')['revenue'].idxmax()
```
