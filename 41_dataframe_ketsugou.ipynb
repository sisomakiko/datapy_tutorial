# 表の結合

```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({'Key': ['k0', 'k1', 'k2'], 
                   'A': ['a0', 'a1', 'a2'],
                   'B': ['b0', 'b1', 'b2']})

df2 = pd.DataFrame({'Key': ['k0', 'k1', 'k2'], 
                   'C': ['c0', 'c1', 'c2'],
                   'D': ['d0', 'd1', 'd2']})
```

```python
df1
```

```python
df2
```

## pd.concat()

```python
# pd.concat([df1, df2, ...], axis=0) 縦結合
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], axis=1)
```

## .merge()

```python
# keyをベースに結合する
df1.merge(df2)
```

```python
df1 = pd.DataFrame({'Key': ['k0', 'k1', 'k2'], 
                   'A': ['a0', 'a1', 'a2'],
                   'B': ['b0', 'b1', 'b2']})

df2 = pd.DataFrame({'Key': ['k0', 'k1', 'k3'], 
                   'C': ['c0', 'c1', 'c3'],
                   'D': ['d0', 'd1', 'd3']})
```

```python
# inner
df1.merge(df2, how='inner', on='Key')
```

```python
# outer
df1.merge(df2, how='outer', on='Key')
```

```python
# left
df1.merge(df2, how='left', on='Key')
```

```python
# right
df1.merge(df2, how='right', on='Key')
```

```python
df1 = pd.DataFrame({'Key': ['k0', 'k1', 'k2'],
                   'ID': ['aa', 'bb', 'cc'],
                   'A': ['a0', 'a1', 'a2'],
                   'B': ['b0', 'b1', 'b2']})

df2 = pd.DataFrame({'Key': ['k0', 'k1', 'k3'], 
                   'ID': ['aa', 'bb', 'cc'],
                   'C': ['c0', 'c1', 'c3'],
                   'D': ['d0', 'd1', 'd3']})
```

```python
df1
```

```python
df2
```

```python
df1.merge(df2, on='Key',suffixes=('_left', '_right'))
```

```python
df_movies = pd.read_csv('archive/tmdb_5000_movies.csv')
df_credits = pd.read_csv('archive/tmdb_5000_credits.csv')
```

```python
df_credits.head()
```

```python
df = df_movies.merge(df_credits, 
                left_on='id', right_on='movie_id', 
                suffixes=('_movies', '_credits'))
df.head()
```
