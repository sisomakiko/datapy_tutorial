# DataFrameとSeriesの便利関数

```python
import numpy as np
import pandas as pd
```

## .unique()　と　nunique()

```python
df_movies = pd.read_csv('archive/tmdb_5000_movies.csv')
df_movies.head()
```

```python
# uniqueな値を取得
df_movies['original_language'].unique()
```

```python
# uniqueな値の数を取得
df_movies['original_language'].nunique()
```

## .value_counts()

```python
# それぞれの値にいくつuniqueな数があるかを取得
df_movies['original_language'].value_counts()
```

## .sort_values()

```python
# sortする（デフォルトは昇順）
df_movies.sort_values('budget')
# 降順は assending=False
```
