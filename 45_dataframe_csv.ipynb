# DataFrameの保存 

```python
import numpy as np
import pandas as pd
```

## .to_csv()

```python
## profit (revenue - budget) の高い順に並べた表を作成してみる
```

```python
# csvを読みこむ
df = pd.read_csv('archive/tmdb_5000_movies.csv')
```

```python
# budget と revenue が0のものを落とす
df = df[~((df['budget'] == 0) | (df['revenue'] == 0))]
```

```python
# profit項目をつくる
df['profit'] = df.apply(lambda row: row['revenue'] - row['budget'], axis=1)
```

```python
# profitでソートする
df = df.sort_values('profit', ascending=False)
df.head()
```

# .to_csv()

```python
df.to_csv('archive/tmdb_5000_movies_profit_sorted.csv', index=False)
```
