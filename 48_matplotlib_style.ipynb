# 装飾 

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-10, 10, 100)
z = 1 / (1 + np.exp(-x))
```

```python
plt.plot(x, z)
```

```python
plt.plot(x, z, 'g--', linewidth=3)
```

```python
plt.plot(x, z, color='violet')
```

```python
plt.plot(x, z, linestyle=':')
```

```python
plt.plot(x, z, 
         marker='o', 
         markersize=10, 
         markerfacecolor='white',
         )
```

```python
#透明度
plt.plot(x, z, 
         alpha=0.5)
```

## 付属情報をつける

```python
plt.plot(x, z, label='sigmoid')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sigmoid(x)')
plt.legend(loc=0)
plt.xticks([-10, -5, 0, 5, 10], rotation=20)
plt.yticks([0, 0.5, 1.0], rotation=45)
```
