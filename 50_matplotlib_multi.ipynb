# 複数のグラフ（図）を描画する

1. plt.subplot()を使う

2-1. plt.figure()を使う

2-2. plt.subplots()を使う

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

## plt.subplot()

```python
x = np.linspace(-10, 10, 100)
plt.subplot(1, 2, 1)
plt.plot(x, x**2)

plt.subplot(1, 2, 2)
plt.plot(x, x**3)
```

## plt.figure()

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

ax1.plot(x, x**2)
ax2.plot(x, x**3)
```

## plt.subplots()

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, x**2)
axes[1].plot(x, x**3)
```

```python
fig, axes = plt.subplots(2, 2)
axes[0, 0].plot(x, x**2)
axes[0, 1].plot(x, x**3)
axes[1, 0].plot(x, x**4)
axes[1, 1].plot(x, x**5)
fig.tight_layout()
```

```python
fig, axes = plt.subplots(1, 2)
axes[0].plot(x, x**2)
axes[0].set_xlabel('xlabel1')
axes[0].set_ylabel('ylabel1')
axes[0].set_title('title1')
axes[1].plot(x, x**3)
axes[1].set_xlabel('xlabel2')
axes[1].set_ylabel('ylabel2')
axes[1].set_title('title2')
fig.tight_layout()
```
