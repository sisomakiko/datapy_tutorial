# plotを保存する

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

#### サイズを指定してplotする

```python
x = np.linspace(-3, 3, 50)
plt.figure(figsize=(10,5))
plt.plot(x, x**2)
```

#### pngファイルで保存： .savefig()

```python
df = pd.read_csv('archive/tmdb_5000_movies.csv')
plt.scatter(df['runtime'], df['budget'], alpha=0.2)
plt.xlabel('runtime')
plt.ylabel('budget')
plt.savefig('runtime_budget_scatter.png')
```

```python
fig, axes = plt.subplots(2, 1, figsize=(10, 4))
axes[0].plot(x, np.cos(x))
axes[0].set_title('cosine')
axes[1].plot(x, np.sin(x))
axes[1].set_title('sine')
fig.tight_layout()
fig.savefig('cosine_sine.png')
```

#### pdfファイルで保存

```python
from matplotlib.backends.backend_pdf import PdfPages

pdf = PdfPages('pdf_files.pdf')
```

```python
fig, axes = plt.subplots()
axes.plot(x, x**2)
pdf.savefig(fig)
pdf.close()
```
